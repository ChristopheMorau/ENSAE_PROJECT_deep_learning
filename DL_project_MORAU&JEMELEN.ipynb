## 2. Training a network model to approximate the score function of the SDE


### 2.1. Definition of the SDE

```python
# in the blog of Song et al., dx = sigma^t * dt
sigma = 20

def std_from_t(t):
  return torch.sqrt((sigma**(2*t) - 1) / (2*np.log(sigma)))

def perturbation(x):
  """ 
  Gives the noisy version of batch image x[i] at random time t[i] according the SDE process.
  returns the perturbed version of all elements of the batch as well as the corresponding respective std
  """

  t = torch.rand(x.shape[0]).to(device)
  std = std_from_t(t)

  z = torch.randn_like(x).to(device)
  perturbed_x = x + z * std[:, None, None, None]
  return perturbed_x, std, t, z
```

### 2.2. Setting up the network model (U-net based)

```python
class GaussianFourierProjection(nn.Module):
  """Gaussian random features for encoding time steps."""  
  def __init__(self, embed_dim, scale=30.):
    super().__init__()
    # Randomly sample weights during initialization. These weights are fixed 
    # during optimization and are not trainable.
    self.W = nn.Parameter(torch.randn(embed_dim // 2) * scale, requires_grad=False)
  def forward(self, x):
    x_proj = x[:, None] * self.W[None, :] * 2 * np.pi
    return torch.cat([torch.sin(x_proj), torch.cos(x_proj)], dim=-1)


class Dense(nn.Module):
  """A fully connected layer that reshapes outputs to feature maps."""
  def __init__(self, input_dim, output_dim):
    super().__init__()
    self.dense = nn.Linear(input_dim, output_dim)
  def forward(self, x):
    return self.dense(x)[..., None, None]


class Model(nn.Module):
  """A time-dependent score-based model built upon U-Net architecture."""

  def __init__(self, channels=[32, 64, 128, 256], embed_dim=256):
    """Initialize a time-dependent score-based network.

    Args:
      marginal_prob_std: A function that takes time t and gives the standard
        deviation of the perturbation kernel p_{0t}(x(t) | x(0)).
      channels: The number of channels for feature maps of each resolution.
      embed_dim: The dimensionality of Gaussian random feature embeddings.
    """
    super().__init__()
    # Gaussian random feature embedding layer for time
    self.embed = nn.Sequential(GaussianFourierProjection(embed_dim=embed_dim),
         nn.Linear(embed_dim, embed_dim))
    # Encoding layers where the resolution decreases
    self.conv1 = nn.Conv2d(3, channels[0], 3, stride=1, bias=False)                                                          
    self.dense1 = Dense(embed_dim, channels[0])
    self.gnorm1 = nn.GroupNorm(4, num_channels=channels[0])
    self.conv2 = nn.Conv2d(channels[0], channels[1], 3, stride=2, bias=False)                                                
    self.dense2 = Dense(embed_dim, channels[1])
    self.gnorm2 = nn.GroupNorm(32, num_channels=channels[1])
    self.conv3 = nn.Conv2d(channels[1], channels[2], 3, stride=2, bias=False)                                               
    self.dense3 = Dense(embed_dim, channels[2])
    self.gnorm3 = nn.GroupNorm(32, num_channels=channels[2])
    self.conv4 = nn.Conv2d(channels[2], channels[3], 3, stride=2, bias=False)
    self.dense4 = Dense(embed_dim, channels[3])
    self.gnorm4 = nn.GroupNorm(32, num_channels=channels[3])    

    # Decoding layers where the resolution increases
    self.tconv4 = nn.ConvTranspose2d(channels[3], channels[2], 3, stride=2, bias=False, output_padding=1)
    self.dense5 = Dense(embed_dim, channels[2])
    self.tgnorm4 = nn.GroupNorm(32, num_channels=channels[2])
    self.tconv3 = nn.ConvTranspose2d(channels[2] + channels[2], channels[1], 3, stride=2, bias=False, output_padding=1)    
    self.dense6 = Dense(embed_dim, channels[1])
    self.tgnorm3 = nn.GroupNorm(32, num_channels=channels[1])
    self.tconv2 = nn.ConvTranspose2d(channels[1] + channels[1], channels[0], 3, stride=2, bias=False, output_padding=1)    
    self.dense7 = Dense(embed_dim, channels[0])
    self.tgnorm2 = nn.GroupNorm(32, num_channels=channels[0])
    self.tconv1 = nn.ConvTranspose2d(channels[0] + channels[0], 3, 3, stride=1)
    
    # The swish activation function
    self.act = lambda x: x * torch.sigmoid(x)
  
  def forward(self, x, t, stds): 
    # Obtain the Gaussian random feature embedding for t   
    embed = self.act(self.embed(t))    
    # Encoding path
    h1 = self.conv1(x)    
    # print('first conv', h1.shape)
    ## Incorporate information from t
    h1 += self.dense1(embed)
    ## Group normalization
    h1 = self.gnorm1(h1)
    h1 = self.act(h1)
    h2 = self.conv2(h1)
    # print('second conv', h2.shape)

    h2 += self.dense2(embed)
    h2 = self.gnorm2(h2)
    h2 = self.act(h2)
    h3 = self.conv3(h2)
    # print('third conv', h3.shape)

    h3 += self.dense3(embed)
    h3 = self.gnorm3(h3)
    h3 = self.act(h3)
    h4 = self.conv4(h3)
    # print('fourth conv',h4.shape)
    h4 += self.dense4(embed)
    h4 = self.gnorm4(h4)
    h4 = self.act(h4)

    # Decoding path
    # print(h4.shape)
    h = self.tconv4(h4)
    ## Skip connection from the encoding path
    h += self.dense5(embed)
    h = self.tgnorm4(h)
    h = self.act(h)
    # print(h.shape)
    # print(h3.shape)
    h = self.tconv3(torch.cat([h, h3], dim=1))
    h += self.dense6(embed)
    h = self.tgnorm3(h)
    h = self.act(h)
    h = self.tconv2(torch.cat([h, h2], dim=1))
    h += self.dense7(embed)
    h = self.tgnorm2(h)
    h = self.act(h)
    h = self.tconv1(torch.cat([h, h1], dim=1))

    # Normalize output
    h = h / stds[:, None, None, None]

    return h

score_model = torch.nn.DataParallel(Model().to(device))
```

### 2.3. Training the network model

_By training the model, we expect to get a network as close as possible to the real score function in terms of output._

```python
def loss_fn(model, x):
  perturbed_x, stds, t, z = perturbation(x)
  score = model(perturbed_x, t, stds)
  loss = torch.mean(torch.sum((score * stds[:, None, None, None] + z)**2, dim=(1,2,3)))
  return loss
```

```python
def optimize(model, lr = 1e-2, n_epochs = 10, data_loader = data_loader, losses = [], clipping = False, max_norm = 1):

  model.train()
  optimizer = Adam(model.parameters(), lr = lr)

  for epoch in trange(n_epochs):
      model.train() 
      avg_loss = 0
      num_items = 0

      for batch_idx, (x, target) in enumerate(data_loader):

          # Clear previous gradients in the graph
          optimizer.zero_grad()
          # Put data and labels on device
          x = x.to(device)
          #compute loss
          loss = loss_fn(model, x)
          # Compute gradients
          loss.backward()
          # Clipping the gradients if option chosen 
          if clipping == True:
            utils.clip_grad_norm(model.parameters(), max_norm)
          # Apply Gradient descent
          optimizer.step()
          
          avg_loss += loss.item() * x.shape[0]
          num_items += x.shape[0]
      print('Average Loss: {:5f}'.format(avg_loss / num_items))
      losses.append(avg_loss / num_items)
  
  return losses
```

>*Once we seem to be stuck at the flat part of the learning curve (after other rounds of Adam epochs the best training loss per epoch is stuck around 33) we reckon that we are not so far from the optimum in the SGD.*

>*To finish the training, we perform a few epochs of the Newton algorithm, which is theoretically faster than SGD under some optimum-closeliness assumptions. The Newton algorithm requires the computation of the inverse Hessian, so given the dimensionality of the problem we rather use an approximation of the Hessian given by the Broyden–Fletcher–Goldfarb–Shanno algorithm (BFGS) using a limited amount of computer memory.*

_Defining optimization of weights through Newton algorithm_

```python
def newton(model, n_epochs = 5, lr = 1e-6, data = data_loader, clipping = False, max_norm = 0.1):

  model.train()
  tqdm_epoch = trange(n_epochs)
  lbfgs = torch.optim.LBFGS(model.parameters(), 
                            lr = lr,
                            max_iter = 10,  
                            history_size = 20)
  def closure():
    lbfgs.zero_grad()
    loss = loss_fn(model, x)
    loss.backward()
    return loss

  newton_losses = []
  c = 1

  for epoch in tqdm_epoch:
    
    avg_loss = 0
    num_items = 0
    
    for x, y in data:
      
      x = x.to(device)   
      if clipping == True:
        utils.clip_grad_norm(model.parameters(), max_norm)
      lbfgs.step(closure)

      avg_loss += loss_fn(score_model, x).item() * x.shape[0]
      num_items += x.shape[0]
    # Print the averaged training loss so far.
    print('Epoch ', c, ' || Avg. loss : ', avg_loss / num_items)
    newton_losses.append(avg_loss / num_items)
    c += 1

  return newton_losses
```

*We see that the loss does not seem to decrease any further with Newton algorithm. It could mean that the optimum of the model is already very close, or that the assumptions necessary to implement Newton's algorithm (notably convexity, lipschitz property and optimum-closeliness) are not satisfied.*

```python
newton_losses = newton(score_model)
```

```python
# Loading the SVHN dataset in numpy format
svhn_data = loadmat("train_32x32.mat") 
images = np.transpose(svhn_data["X"], (3, 0, 1, 2))
# Convert the images to the LAB color space
images_lab = color.rgb2lab(images)
```

```python
# Compute the color histograms for each image
def compute_histograms(images):
    histograms = []
    for image in images:
        hist, _ = np.histogram(image, bins=256, range=(-128, 128))
        histograms.append(hist)
    histograms = np.array(histograms)
    return histograms

histograms = compute_histograms(images_lab)
# Normalize the histograms
histograms_norm = histograms / np.sum(histograms, axis=1, keepdims=True)
```

## 3. Using the score approximation of the trained network to generate samples

### 3.1. Definition of Euler-Maruyama sampler (discretized reverse SDE solver)

```python
def get_x_delta(x, delta, t, score):
  z = torch.randn(x.shape).to(device)

  x_to_return = x + sigma**(2*t)[:, None, None, None] *score * delta 
  x_delta = x_to_return + sigma**t[:, None, None, None]*math.sqrt(delta)*z
  t_delta = t - delta*torch.ones(num_samples, device=device)

  return x_delta, x_to_return, t_delta[0]



def euler_Maruyama_sampler(score_model, num_samples, delta):

  #draw from the prior (law of )
  shape_x = (num_samples, 3, 32, 32) # shape of our SVHN data
  samples_from_prior = torch.randn(shape_x).to(device) * std_from_t(torch.ones(64, device=device))[:, None, None, None]

  t_delta = 1 - delta
  x_delta = samples_from_prior
  num_iter = int((1 - 1e-3) // delta)
  with torch.no_grad():
    for i in trange(num_iter):
      t_tensor = t_delta*torch.ones(num_samples, device=device)
      score = score_model(x_delta, t_tensor, std_from_t(t_tensor))
      x_delta, x_to_return, t_delta = get_x_delta(x_delta, delta, t_tensor, score)
  return x_to_return
  
```

### 3.2. Definition of Langevin corrector et predictor-corrector (PC) sampler
_Euler Maruyama sampler is a more sophisticated MCMC sampler with correction of time steps to improve the sampling_

```python
def langevin(x_delta, epsilon, score):
  x = x_delta + epsilon * score + torch.sqrt(2 * epsilon) * torch.randn_like(x_delta)
  return x 
```

```python
def pc_sampler(score_model, num_samples, delta, langevin_score_importance):

  #draw from the prior (law of )
  shape_x = (num_samples, 3, 32, 32)
  samples_from_prior = torch.randn(shape_x).to(device) * std_from_t(torch.ones(64, device=device))[:, None, None, None]

  t_delta = 1 - delta
  x_delta = samples_from_prior
  num_iter = int((1 - 1e-3) // delta)
  with torch.no_grad():
    for i in trange(num_iter):
      t_tensor = t_delta*torch.ones(num_samples, device=device)
      score = score_model(x_delta, t_tensor, std_from_t(t_tensor))

      #PREDICTOR
      #sample with euler-maruyama
      x_delta, x_to_return, t_delta = get_x_delta(x_delta, delta, t_tensor, score)


      #CORRECTOR:
      #We use the same method as the tutorial to determine the Langevin step
      score_norm = torch.norm(score.reshape(score.shape[0], -1), dim=-1).mean()
      noise_norm = np.sqrt(np.prod(x_delta.shape[1:]))
      epsilon = 2 * (langevin_score_importance * noise_norm / score_norm)**2

      #We apply Langevin MCMC to correct x_delta
      x_delta = langevin(x_delta, epsilon, score)
      
  return x_to_return
```
